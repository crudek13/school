```python
import pandas as pd
import numpy as np
import sqlite3
import warnings
warnings.filterwarnings("ignore")
```

```python
s = pd.read_excel ("startups.xlsx")
```

```python
s['month'] = s['valuation_date'].dt.month
```

```python
f1 = s['country'].value_counts().rename_axis('country').reset_index(name='counts')
f2 = s['industry'].value_counts().rename_axis('industry').reset_index(name='counts')
f3 = s['month'].value_counts().rename_axis('month').reset_index(name='counts')
```

```python
s1 = s.groupby("country").agg({"valuation": "sum"})
s2 = s.groupby("industry").agg({"valuation": "sum"})
s3 = s.groupby("month").agg({"valuation": "sum"})
```

```python
db_conn = sqlite3.connect("st_DB.db")
```

```python
f1.to_sql("f1_DB", db_conn, if_exists="replace")
f2.to_sql("f2_DB", db_conn, if_exists="replace")
f3.to_sql("f3_DB", db_conn, if_exists="replace")
s1.to_sql("s1_DB", db_conn, if_exists="replace")
s2.to_sql("s2_DB", db_conn, if_exists="replace")
s3.to_sql("s3_DB", db_conn, if_exists="replace")
s.to_sql("s_DB", db_conn, if_exists="replace")
```

```python
s.to_sql("s_DB", db_conn, if_exists="replace")
```
