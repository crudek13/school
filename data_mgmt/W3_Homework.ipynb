```python
#2.B
mpg['make'] = mpg.carname.str.split().str.get(0)
mpg['model'] = mpg.carname.str.split().str.get(1)
mpg.rename(columns={'model year':'year'}, inplace = True)
# year = 70, model = 1200d, make = hi
mpg_sm = mpg.sort_values(by=['mpg'], ascending = False).tail(1)
print(mpg_sm[['model_year', 'make', 'model']])
```

```python
#2.c
m2c = mpg.agg({'weight': ['mean','min','max'],
              'mpg': ['mean','min','max']})

print(m2c.T)
```

```python
#3
mpg['horsepower'] = pd.to_numeric(mpg['horsepower'],errors='coerce')
mpg

m3a = mpg[['make', 'horsepower']]
m3a = m3a.groupby(['make'],sort=True).mean()
m3a.sort_values(by=['horsepower'], ascending = False)

mpg.groupby(["make"])[["horsepower"]].describe()

#at observation 258 it hits the 50% mark
mpg_cumu = mpg.sort_values(by=['horsepower'], ascending = True)
mpg_cumu['hp_pct'] = mpg_cumu['horsepower']/mpg_cumu['horsepower'].sum()
mpg_cumu['hp_cum'] = mpg_cumu['horsepower'].cumsum(axis = 0)
mpg_cumu['hp_cum_pct'] = mpg_cumu['hp_cum']/mpg_cumu['horsepower'].sum()
mpg_cumu_f = mpg_cumu[['horsepower', 'hp_cum', 'hp_pct', 'hp_cum_pct']]
#print(mpg_cumu_f.to_string()) -- way to see whole data frame
new_cumu_f = mpg_cumu_f.loc[mpg_cumu_f['hp_cum_pct'] > .5]
new_cumu_f
```

```python
#Q5
m = np.array ([[10,20,30], [40,50,60], [70, 71,72]])

a1 = pd.DataFrame(m, columns=['CO1', 'CO2', 'CO3'])
a1.rename(index={0: "RO1", 1: "RO2", 2: "RO3"})

a1['t_day'] = pd.to_datetime('today')

a1['Today_Year'] = a1['t_day'].dt.year
a1['Today_Month'] = a1['t_day'].dt.month
a1['Today_week'] = a1['t_day'].dt.week
a1['Today_day'] = a1['t_day'].dt.day
wd={0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
a1['Day_of_week_name']=a1['t_day'].dt.weekday.map(wd)
a1['Today_hour'] = a1['t_day'].dt.hour
a1['Today_minute'] = a1['t_day'].dt.minute
```
